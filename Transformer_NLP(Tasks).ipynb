## Reading a gdoc file from the drive

```python
from transformers import pipeline

# Read the contents of the file
with open("/content/drive/MyDrive/Colab Notebooks/HAMOYE_LLM/Promise Alx Journey .gdoc", "r") as file:

    context = file.read()

# Initialize the question-answering pipeline
qan = pipeline('question-answering')

# Ask a question based on the context from the file
question = "What is the main topic of the document?"
result = qan(question=question, context=context)

# Print the answer
print(result)

```

## Reading from a url

```python
import requests
from transformers import pipeline

# URL of the document on the web
url = 'https://docs.google.com/document/d/1lW0fLdgZbvgaOtqWTECPnWsBa7YL5glhQWfkiaqFeJQ/edit?usp=drive_link'

# Download the contents of the document
response = requests.get(url)
context = response.text

# Initialize the question-answering pipeline
qan = pipeline('question-answering')

# Ask a question based on the web document
question = "What is the main topic of the document?"
result = qan(question=question, context=context)

# Print the answer
print(result)
```
